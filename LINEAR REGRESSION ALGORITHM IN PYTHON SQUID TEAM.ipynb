```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
# from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
```

```python
df = pd.read_csv("SeoulBikeData.csv")
df.head()
```

```python
y = df["Rented Bike Count"]
x = df.loc[:, df.columns != "Rented Bike Count"]
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
x["Date"] = number.fit_transform(x["Date"].astype("str"))

x.head()                                      
```

```python
number = LabelEncoder()
x["Seasons"] = number.fit_transform(x["Seasons"].astype("str"))

x.head()  
```

```python
number = LabelEncoder()
x["Holiday"] = number.fit_transform(x["Holiday"].astype("str"))

x.head() 
```

```python
number = LabelEncoder()
x["Functioning Day"] = number.fit_transform(x["Functioning Day"].astype("str"))

x.head() 
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
RentedBikeCount_pred = model.predict(X_test)
```

```python
#coefficients

print("Coefficients: \n", model.coef_)

print("Intercept: \n", model.intercept_)

#mean squared error

print("Mean squared error: %.2f"
      % mean_squared_error(y_test, RentedBikeCount_pred))

#coefficient of determination: 1 is perfect prediction

print("Coefficient of determination: %.2f"
      % r2_score(y_test, RentedBikeCount_pred))

#mean absolute error

print("Mean absolute error: %.2f"
     %mean_absolute_error(y_test, RentedBikeCount_pred))
```

```python
mse = mean_squared_error(y_test, RentedBikeCount_pred)
rmse = np.sqrt(mse)
print("RMSE: %.2f" %rmse)
```
